# Contexto

Un hospital de la ciudad especializado en salud mental está intentando iniciar una investigación especial sobre la enfermedad de Alzheimer. El hospital tiene un conjunto de datos que contiene amplia información sanitaria de los pacientes, cada uno identificado de forma única con un identificador. El conjunto de datos incluye detalles demográficos, factores de estilo de vida, historial médico, mediciones clínicas, evaluaciones cognitivas y funcionales, síntomas y un diagnóstico de la enfermedad de Alzheimer. Este análisis tiene como objetivo probar el concepto antes de comenzar formalmente la investigación. Los médicos están interesados en explorar los factores asociados con el Alzheimer y desarrollar modelos predictivos, para ver si realmente es posible predecir el resultado basándose en los datos clínicos disponibles.

# Información del conjunto de datos:

## Paciente:

- PatientID: Un identificador único asignado a cada paciente

- Gender: Sexo de los pacientes, donde 0 representa masculino y 1 representa femenino.

- Ethnicity: La etnicidad de los pacientes, codificada de la siguiente manera: 0: Caucásico, 1: Afroamericano, 2: Asiático, 3: Otro

- EducationLevel: El nivel de educación de los pacientes, codificado de la siguiente manera: 0: Ninguno, 1: Secundaria, 2: Licenciatura, 3: Superior

- BMI: Índice de masa corporal de los pacientes

- Smoking: Booleano

- AlcoholConsumption: Consumo semanal de alcohol en unidades

- PhysicalActivity: Actividad física semanal en horas, que va de 0 a 10.

- DietQuality: Puntuación de calidad de la dieta

- SleepQuality: Puntuación de calidad del sueño

- FamilyHistoryAlzheimers: Antecedentes familiares de enfermedad de Alzheimer, donde 0 indica No y 1 indica Sí.

- CardiovascularDisease: Presencia de enfermedad cardiovascular, donde 0 indica No y 1 indica Sí.

- Diabetes: presencia de diabetes, donde 0 indica no y 1 indica sí.

- Depresión: presencia de depresión, donde 0 indica no y 1 indica sí.

- Traumatismo craneal: antecedentes de traumatismo craneal, donde 0 indica no y 1 indica sí.

- Hipertensión: presencia de hipertensión, donde 0 indica no y 1 indica sí.

## Mediciones clínicas:

- PA sistólica: presión arterial sistólica, que oscila entre 90 y 180 mmHg.

- PA diastólica: presión arterial diastólica, que oscila entre 60 y 120 mmHg.

- Colesterol total: niveles de colesterol total, que oscilan entre 150 y 300 mg/dl.

- Colesterol LDL: niveles de colesterol de lipoproteínas de baja densidad, que oscilan entre 50 y 200 mg/dl.

- Colesterol HDL: niveles de colesterol de lipoproteínas de alta densidad, que oscilan entre 20 y 100 mg/dl.

- ColesterolTriglicéridos: Niveles de triglicéridos, que van de 50 a 400 mg/dL.

## Evaluaciones cognitivas y funcionales:

- MMSE: Puntaje del Mini-Mental State Examination, que va de 0 a 30. Los puntajes más bajos indican deterioro cognitivo.

- Evaluación funcional: Puntaje de la evaluación funcional, que va de 0 a 10. Los puntajes más bajos indican un mayor deterioro.

- Quejas de memoria: Presencia de quejas de memoria, donde 0 indica No y 1 indica Sí.

- Problemas de conducta: Presencia de problemas de conducta, donde 0 indica No y 1 indica Sí.

- AVD: Puntaje de Actividades de la vida diaria, que va de 0 a 10. Los puntajes más bajos indican un mayor deterioro.

- Confusión: Booleano

- Desorientación: Booleano

- Cambios de personalidad: Booleano

- Dificultad para completar tareas: Booleano

- Olvido: Booleano

## Información de diagnóstico:

- Diagnóstico: Estado de diagnóstico de la enfermedad de Alzheimer, donde 0 indica No y 1 indica Sí.

- DoctorInCharge: Información confidencial (médico responsable) [No es importante para el analisis]
